```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
match_res_path = "ipl_matches_2008_2022.csv"
match_res = pd.read_csv(match_res_path)
```

```python
ball_by_ball_path = "ipl_ball_by_ball_2008_2022.csv"
ball_by_ball = pd.read_csv(ball_by_ball_path)
```

```python
match_res
```

```python
ball_by_ball.head()
```

```python
total_runs = ball_by_ball.groupby(['id','innings']).sum()['total_run'].add(1).reset_index()
```

```python
total_runs.head()
```

```python
total_runs = total_runs[total_runs['innings']==1].reset_index()
total_runs.head()
```

```python
match_df = match_res.merge(total_runs[['id','total_run']],left_on='id',right_on='id')
match_df
```

```python
match_df = match_df[['id','venue','team1','team2','total_run','winning_team']]
match_df
```

```python
second_innings = match_df.merge(ball_by_ball,on='id')
second_innings
```

```python
second_innings = second_innings[second_innings['innings']==2]
second_innings
```

```python
second_innings['batting_team'].value_counts()
```

```python
import warnings
warnings.filterwarnings("ignore")

second_innings['batting_team'] = second_innings['batting_team'].replace('Delhi Daredevils','Delhi Capitals')
second_innings['batting_team'] = second_innings['batting_team'].replace('Deccan Chargers','Sunrisers Hyderabad')
second_innings['batting_team'] = second_innings['batting_team'].replace('Punjab Kings','Kings XI Punjab')
```

```python
second_innings['batting_team'].value_counts()
```

```python
teams = [
    'Delhi Capitals',
    'Kolkata Knight Riders',
    'Royal Challengers Bangalore',
    'Mumbai Indians',
    'Sunrisers Hyderabad',
    'Rajasthan Royals',
    'Chennai Super Kings',
    'Kings XI Punjab',
    'Gujarat Titans',
    'Lucknow Super Giants']
    
```

```python
second_innings = second_innings[second_innings['batting_team'].isin(teams)]
second_innings['batting_team'].value_counts()
```

```python
second_innings['venue'].value_counts()
```

```python
second_innings['current_score'] = second_innings.groupby('id')['total_run_y'].cumsum()
second_innings
```

```python
second_innings.head(10)
```

```python
second_innings['required_runs'] = second_innings['total_run_x'] - second_innings['current_score']
second_innings
```

```python
second_innings['balls_left'] = 120 - (second_innings['overs']*6 + second_innings['ball_number'])
second_innings
```

```python
wickets = second_innings.groupby('id')['iswicket_delivery'].cumsum()
second_innings['wickets_remaining'] = 10 - wickets
second_innings.head(10)
```

```python
second_innings['crr'] = round(second_innings['current_score']*6/(120 - second_innings['balls_left']),2)
second_innings['req_rr'] = round((second_innings['total_run_x'] - second_innings['current_score'])*6/second_innings['balls_left'],2)
```

```python
second_innings
```

```python
def winner(team):
    return 1 if team['batting_team'] == team['winning_team'] else 0
```

```python
second_innings['results'] = second_innings.apply(winner,axis=1)
second_innings
```

```python
second_innings['bowling_team'] = second_innings.apply(lambda row: row['team1'] if row['batting_team'] == row['team2'] else row['team2'], axis=1)
second_innings
```

```python
data = [
    'venue',
    'batting_team',
    'bowling_team',
    'required_runs',
    'balls_left',
    'wickets_remaining',
    'total_run_x',
    'crr',
    'req_rr',
    'results']
```

```python
required_data = second_innings[data]
final_data = required_data.sample(required_data.shape[0])
```

```python
final_data.dropna(inplace=True)
```

```python
final_data = final_data[final_data['balls_left']!=0]
```

```python
final_data.to_csv('final_data.csv',index=False)
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x = final_data.iloc[:,:-1]
y = final_data.iloc[:,-1]

x
```

```python
train_x,test_x,train_y,test_y = tts(x,y,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(drop='first'),['venue','batting_team','bowling_team'])
    ],
    remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

pipe = Pipeline([
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(train_x,train_y)
```

```python
pred_y = pipe.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_y,pred_y)
```

```python
pipe.predict_proba(test_x)[0]
```

```python
import pickle
with open('pipe.pkl', 'wb') as f:
    pickle.dump(pipe, f)
```
